```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize= (5, 5))

plt.scatter(x, y, color= "green", edgecolors= "black", linewidths= 2)

plt.title("Scatter Plot with Edge")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```

```python
plt.figure(figsize= (5, 5))

plt.scatter(x, y, color= "yellow", edgecolors= "black", linewidths= 2)

plt.title("Scatter Plot with Edge")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```

```python
plt.figure(figsize= (5, 5))

plt.scatter(x, y, color= "c", edgecolors= "black", linewidths= 12)

plt.title("Scatter Plot with Edge")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```

```python
plt.figure(figsize= (5, 5))

plt.scatter(x, y, color= "y", edgecolors= "black", linewidths= 4)

plt.title("Scatter Plot with Edge")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```

```python
z = np.random.uniform(0, 1, 100)
```

```python
z
```

```python
plt.figure(figsize= (5, 5))

cmap = plt.cm.get_cmap("viridis")

plt.scatter(x, y, c= z, cmap= cmap, edgecolors= ["r", "g"], linewidths= [1, 2])
plt.colorbar()

plt.title("Scatter Plot with Edge")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```

```python
plt.figure(figsize= (5, 5))

cmap = plt.cm.get_cmap("viridis")

plt.scatter(x, y, c= z, cmap= cmap, edgecolors= ["r", "g", "y"], linewidths= [2, 3, 5])
plt.colorbar()

plt.title("Scatter Plot with Edge")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```

```python
plt.figure(figsize= (5, 5))

cmap = plt.cm.get_cmap("viridis")

plt.scatter(x, y, c= z, cmap= cmap, edgecolors= ["r", "g", "y", "b"], linewidths= [2, 3, 2, 2])
plt.colorbar()

plt.title("Scatter Plot with Edge")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
```
