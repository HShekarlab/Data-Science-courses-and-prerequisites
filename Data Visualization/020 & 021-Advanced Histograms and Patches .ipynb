```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sample_men = [112, 90, 76, 94, 50, 70, 113, 140, 150, 160, 139, 120, 76, 77, 99, 88, 110, 67, 50, 80]
blood_sample_women = [132, 80, 66, 64, 90, 80, 143, 150, 130, 140, 119, 60, 76, 95, 69, 128, 150, 167, 150, 180]
```

```python
plt.figure(figsize= (5, 5))

plt.hist([blood_sample_men, blood_sample_women], 
          bins= [80, 100, 125, 150],
          color= ["green", "orange"],
          label= ["men", "women"])

plt.legend()
```

```python
plt.figure(figsize= (5, 5))

plt.hist([blood_sample_men, blood_sample_women], 
          bins= [80, 100, 125, 150],
          color= ["green", "orange"],
          label= ["men", "women"])

plt.legend()

plt.title("Blood Sugar Chart")
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
```

```python
plt.figure(figsize= (5, 5))

plt.hist([blood_sample_men, blood_sample_women], 
          bins= [80, 100, 125, 150],
          color= ["green", "orange"],
          rwidth= 1,
          label= ["men", "women"])

plt.legend()

plt.title("Blood Sugar Chart")
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
```

***Patches in Histogram***

```python
plt.figure(figsize= (5, 5))

n, bins, patchs= plt.hist([blood_sample_men, blood_sample_women], 
                           bins= [80, 100, 125, 150],
                           color= ["green", "orange"],
                           rwidth= 0.95,
                           label= ["men", "women"])

plt.legend()

plt.title("Blood Sugar Chart")
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
```
