***Changing fill color of different areas (negative vs positive for example)***

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [-1, -4, -3, 0, 4, 6, 7]
y = [-4, -5, 2, -1, 6, 4, -2]

x = np.array(x)
y = np.array(y)

plt.figure(figsize= (5, 5))

plt.plot(x, y)

plt.xticks([x for x in np.arange(-8, 8, 1)])
plt.yticks([x for x in np.arange(-8, 8, 1)])
```

```python
x = [-1, -4, -3, 0, 4, 6, 7]
y = [-4, -5, 2, -1, 6, 4, -2]

x = np.array(x)
y = np.array(y)

plt.figure(figsize= (5, 5))

plt.plot(x, y)

plt.xticks([x for x in np.arange(-8, 8, 1)])
plt.yticks([x for x in np.arange(-8, 8, 1)])

plt.fill_between(x, y, where= (x < 0), color= "m", alpha= 0.7)
```
