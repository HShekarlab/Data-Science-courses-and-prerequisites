***Adding Legend***

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [20, 15, 35, 10, 25]
my_labels = ["Sour cherry", "Sour orange", "Watermelon", "Olive", "Persimmon"]

plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels)
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels)

plt.legend()
```

***Adding Title in piechart and Title in Legend***

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels)

plt.title("Fruite Data")
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels)

plt.title("Fruite Data")

plt.legend(title= "Fruites Name")
```

***Adding percnetage***

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, autopct= "%1.1f%%")
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, autopct= "%1.2f%%")
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, autopct= "%1.3f%%")
```

***Roating a pie chart***

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, autopct= "%1.1f%%", startangle= 45)
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, autopct= "%1.1f%%", startangle= 100)
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, autopct= "%1.1f%%")

plt.title("Fruite Data")

plt.legend(title= "Fruites Name", loc= "lower left")
```
