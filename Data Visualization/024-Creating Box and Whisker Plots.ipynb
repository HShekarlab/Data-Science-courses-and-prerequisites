```python
x_2 = [10, 20, 30, 40, 50, 60, 70, 120]
x_3 = [10, 20, 60, 40, 50, 30, 90, 130]

plt.figure(figsize= (4, 4))

plt.boxplot(x_2,
            widths= 0.2,
            labels= ["UK"],
            showmeans= True,
            sym= "g+",
            boxprops= dict(color= "g"),
            capprops= dict(color= "b"),
            whiskerprops= dict(color= "m"))

plt.boxplot(x_3,
            widths= 0.2,
            labels= ["UK"],
            showmeans= True,
            sym= "c*",
            boxprops= dict(color= "r"),
            capprops= dict(color= "y"),
            whiskerprops= dict(color= "k"))
```

```python
x_2 = [10, 20, 30, 40, 50, 60, 70, 120]
x_3 = [10, 20, 60, 40, 50, 30, 90, 130]

y = [x_2, x_3]

plt.figure(figsize= (4, 4))

plt.boxplot(y,
            widths= 0.2,
            labels= ["Python", "C++"],
            showmeans= True,
            sym= "c+",
            boxprops= dict(color= "r"),
            capprops= dict(color= "b"),
            whiskerprops= dict(color= "m"))
```

```python
x_2 = [10, 20, 30, 40, 50, 60, 70, 120]
x_3 = [10, 20, 60, 40, 50, 30, 90, 130]
x_4 = [10, 20, 40, 50, 70, 90, 130]

y = [x_2, x_3, x_4]

plt.figure(figsize= (4, 4))

plt.boxplot(y,
            widths= 0.2,
            labels= ["Python", "C++", "R"],
            showmeans= True,
            sym= "c+",
            boxprops= dict(color= "r"),
            capprops= dict(color= "b"),
            whiskerprops= dict(color= "m"))
```
