```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = ["A", "B", "C", "D"]

y_1 = np.array([10, 20, 10, 30])
y_2 = np.array([20, 25, 15, 25])
y_3 = np.array([12, 15, 19, 6])
y_4 = np.array([10, 29, 13, 19])
```

```python
plt.figure(figsize= (5, 5))

plt.bar(x, y_1, color= "orange")

plt.title("Scores by Teams in 4 Rounds in UK")
plt.xlabel("Team")
plt.ylabel("Score")

plt.legend("Round 1")
```

```python
plt.figure(figsize= (5, 5))

plt.bar(x, y_1, color= "orange")
plt.bar(x, y_2, bottom= y_1 , color= "c")

plt.title("Scores by Teams in 4 Rounds in UK")
plt.xlabel("Team")
plt.ylabel("Score")

plt.legend(["Round 1", "Round 2"])
```

```python
plt.figure(figsize= (5, 5))

plt.bar(x, y_1, color= "orange")
plt.bar(x, y_2, bottom= y_1 , color= "c")
plt.bar(x, y_3, bottom= y_1 + y_2, color= "m")

plt.title("Scores by Teams in 4 Rounds in UK")
plt.xlabel("Team")
plt.ylabel("Score")

plt.legend(["Round 1", "Round 2", "Round 3"])
```

```python
plt.figure(figsize= (5, 5))

plt.bar(x, y_1, color= "orange")
plt.bar(x, y_2, bottom= y_1 , color= "c")
plt.bar(x, y_3, bottom= y_1 + y_2, color= "m")
plt.bar(x, y_4, bottom= y_1 + y_2 + y_3, color= "b")

plt.title("Scores by Teams in 4 Rounds in UK")
plt.xlabel("Team")
plt.ylabel("Score")

plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
```

```python
plt.figure(figsize= (5, 5))

plt.barh(x, y_1, color= "orange")
plt.barh(x, y_2, left= y_1 , color= "c")
plt.barh(x, y_3, left= y_1 + y_2, color= "m")
plt.barh(x, y_4, left= y_1 + y_2 + y_3, color= "b")

plt.title("Scores by Teams in 4 Rounds in UK")
plt.xlabel("Team")
plt.ylabel("Score")

plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
```
