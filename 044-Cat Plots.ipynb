```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.catplot(data= tips, x= "day", y= "tip", hue= "sex", kind= "swarm", palette= "husl")

plt.title("Tip by Day and Sex")
plt.xlabel("Day of the Week")
plt.ylabel("Tip")
```

```python
sns.catplot(data= tips, x= "day", y= "tip", hue= "sex", kind= "point", palette= "Set2")

plt.title("Tip by Day and Sex")
plt.xlabel("Day of the Week")
plt.ylabel("Tip")
```

```python
sns.catplot(data= tips, x= "day", y= "tip", hue= "sex", kind= "bar", palette= "Paired")

plt.title("Tip by Day and Sex")
plt.xlabel("Day of the Week")
plt.ylabel("Tip")
```

```python
sns.catplot(data= tips, x= "day", y= "tip", hue= "sex", kind= "box", palette= "viridis")

plt.title("Tip by Day and Sex")
plt.xlabel("Day of the Week")
plt.ylabel("Tip")
```

```python
sns.catplot(data= tips, x= "day", y= "tip", hue= "sex", kind= "violin", palette= "flare")

plt.title("Tip by Day and Sex")
plt.xlabel("Day of the Week")
plt.ylabel("Tip")
```
