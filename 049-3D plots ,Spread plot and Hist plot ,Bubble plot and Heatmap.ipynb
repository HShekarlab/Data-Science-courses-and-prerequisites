***3D Plot***

```python
np.random.seed(123)

df = pd.DataFrame({"x": np.random.randn(100),
                   "y": np.random.randn(100),
                   "z": np.random.randn(100)})

fig = px.scatter_3d(df, x= "x", y= "y", z= "z")
fig.show()
```

***Spread Plot and Hist Plot***

***Bubble Plot and Heatmap***

```python
df_1 = pd.DataFrame({"x": np.random.rand(50),
                     "y": np.random.rand(50),
                     "size": np.random.rand(50),
                     "color": np.random.randint(1, 4, size= 50)})

df_1
```

```python
fig = px.scatter(df_1, x= "x", y= "y", size= "size", color= "color")
fig.show()
```

```python
heatmap_data = df_1.pivot_table(values= "size", index= "x", columns= "y")

fig = px.imshow(heatmap_data, color_continuous_scale= "Blues")
fig.show()
```

```python
fig = px.imshow(heatmap_data, color_continuous_scale= "matter")
fig.show()
```

```python
fig = px.imshow(heatmap_data, color_continuous_scale= "magma")
fig.show()
```

```python
fig = px.imshow(heatmap_data, color_continuous_scale= "cividis")
fig.show()
```

```python
fig = px.imshow(heatmap_data, color_continuous_scale= "twilight")
fig.show()
```
