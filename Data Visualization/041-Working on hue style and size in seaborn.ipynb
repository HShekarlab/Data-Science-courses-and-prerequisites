```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
sns.scatterplot(data= iris, x= "sepal_length", y= "petal_width")
```

```python
sns.scatterplot(data= iris, x= "sepal_length", y= "petal_width", hue= "species")
```

```python
sns.scatterplot(data= iris, 
                x= "sepal_length", y= "petal_width", 
                hue= "species",
                style= "species")
```

```python
sns.scatterplot(data= iris, 
                x= "sepal_length", y= "petal_width", 
                hue= "species",
                style= "species",
                size= "sepal_width")
```

```python
sns.scatterplot(data= iris, 
                x= "sepal_length", y= "petal_width", 
                hue= "species",
                style= "species",
                size= "sepal_width",
                sizes= (20, 200))
```

```python
sns.scatterplot(data= iris, 
                x= "sepal_length", y= "petal_width", 
                hue= "species",
                style= "species",
                palette= "rocket",
                size= "sepal_width",
                sizes= (20, 200))
```
