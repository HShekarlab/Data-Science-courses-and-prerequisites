# ***Types of Charts***

**plot(x, y)**

```python
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, linewidth=2)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, linewidth=2)

plt.show()
```

**scatter(x, y)**

```python
np.random.seed(10)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y, vmin=0, vmax=100)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**bar(x, height)**

```python
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**stem(x, y)**

```python
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

fig, ax = plt.subplots(figsize=(5, 3))
ax.stem(x, y)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**fill_between(x, y1, y2)**

```python
np.random.seed(5)
x = np.linspace(0, 8, 16)
y_1 = 3 + 4 * x / 8 + np.random.uniform(0, 0.5, len(x))
y_2 = 1 + 2 * x / 8 + np.random.uniform(0, 0.5, len(x))

fig, ax = plt.subplots(figsize=(5, 3))
ax.fill_between(x, y_1, y_2, alpha=0.5, linewidth=0)
ax.plot(x, (y_1 + y_2) / 2, linewidth=2)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**stackplot(x, y)**

```python
x = np.arange(0, 10, 2)
a_y = [1, 1.25, 2, 2.75, 3]
b_y = [1, 1, 1, 1, 1]
c_y = [2, 1, 2, 1, 2]
y = np.vstack([a_y, b_y, c_y])

fig, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(x, y)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**stairs(values)**

```python
x = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

fig, ax = plt.subplots(figsize=(5, 3))
ax.stairs(x, linewidth=2.5)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**hist(x)**

```python
np.random.seed(10)
x = 4 + np.random.normal(0, 1.5, 20)

fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**boxplot(X)**

```python
np.random.seed(10)
x = np.random.normal((3, 5, 4), (1.25, 1, 1.25), (100, 3))

fig, ax = plt.subplots(figsize=(5, 3))
X = ax.boxplot(x, positions=[2, 4, 6], widths=1.5, patch_artist=True, showmeans=False, showfliers=False,
               medianprops= {"color": "white", "linewidth": 0.5},
               boxprops= {"facecolor": "C0", "edgecolor": "white", "linewidth": 0.5},
               whiskerprops= {"color": "C0", "linewidth": 1.5},
               capprops= {"color": "C0", "linewidth": 1.5})
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**errorbar(x, y, yerr, xerr)**

```python
np.random.seed(1)
x = [2, 4, 6]
y = [3.5, 5, 4.2]
y_err = [0.9, 1.2, 0.5]

fig, ax = plt.subplots(figsize=(5, 3))
ax.errorbar(x, y, y_err, fmt= "o", linewidth=2, capsize=6)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**violinplot(D)**

```python
np.random.seed(10)
x = np.random.normal((3, 5, 4), (0.75, 1, 0.75), (200, 3))

fig, ax = plt.subplots(figsize=(5, 3))
X = ax.violinplot(x, [2, 4, 6], widths=2, showmeans=False, showmedians=False, showextrema=False)
for item in X['bodies']:
    item.set_alpha(0.9)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**eventplot(D)**

```python
np.random.seed(1)
x = [2, 4, 6]
X = np.random.gamma(4, size=(3, 50))

fig, ax = plt.subplots(figsize=(5, 3))
ax.eventplot(X, orientation="vertical", lineoffsets=x, linewidth= 0.75)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

**plot_surface(X, Y, Z)**

```python
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2)
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

**plot_trisurf(x, y, z)**

```python
n_radii = 8
n_angles = 36

radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2 * np.pi, n_angles, endpoint=False)[..., np.newaxis]

x = np.append(0, (radii * np.cos(angles)).flatten())
y = np.append(0, (radii * np.sin(angles)).flatten())
z = np.sin(- x * y)
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.plot_trisurf(x, y, z, vmin=z.min() * 2)
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

**voxels([x, y, z], filled)**

```python
x, y, z = np.indices((8, 8, 8))

cube1 = (x < 3) & (y < 3) & (z < 3)
cube2 = (x >= 5) & (y >= 5) & (z >= 5)

voxelarray = cube1 | cube2

fig, ax = plt.subplots(subplot_kw = {"projection": "3d"})
ax.voxels(voxelarray, edgecolor='k')
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
