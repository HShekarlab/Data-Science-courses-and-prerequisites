```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.histplot(data= tips, x= "total_bill", kde= True, color= "c", alpha= 0.7)
```

```python
sns.scatterplot(data= tips, x= "total_bill", y= "tip", hue= "time")
```

```python
sns.scatterplot(data= tips, x= "total_bill", y= "tip", hue= "day", palette= "viridis")
```

```python
sns.scatterplot(data= tips, x= "total_bill", y= "tip", hue= "size", palette= "flare")
```

```python
fig, axes= plt.subplots(nrows= 1, ncols= 2, figsize= (10, 5))

sns.histplot(ax= axes[0], data= tips, x= "total_bill", kde= True, color= "c", alpha= 0.7)

axes[0].set_title("Distribution of Total Bill")
axes[0].set_xlabel("Total Bill")
axes[0].set_ylabel("Count")

sns.scatterplot(ax= axes[1], data= tips, x= "total_bill", y= "tip", hue= "time", palette= "magma")

axes[0].set_title("Total Bill vs Tip")
axes[0].set_xlabel("Total Bill")
axes[0].set_ylabel("Tip")
```
