```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y_1 = np.sin(x)
y_1
```

```python
y_2 = np.cos(x)
y_2
```

```python
y_3 = np.tan(x)
y_3
```

```python
fig_1, axes_1 = plt.subplots(figsize= (4, 4))
axes_1.plot(x, y_1)
axes_1.set_title("Sin")

fig_2, axes_2 = plt.subplots(figsize= (8, 5))
axes_2.plot(x, y_2)
axes_2.set_title("Cosin")

fig_3, axes_3 = plt.subplots(figsize= (10, 10))
axes_3.plot(x, y_3)
axes_3.set_title("Tangent")
```

```python
fig_1, axes_1 = plt.subplots(figsize= (4, 4))
axes_1.plot(x, y_1, "g+")
axes_1.set_title("Sin")

fig_2, axes_2 = plt.subplots(figsize= (8, 5))
axes_2.plot(x, y_2, "b*")
axes_2.set_title("Cosin")

fig_3, axes_3 = plt.subplots(figsize= (10, 10))
axes_3.plot(x, y_3, "m-.")
axes_3.set_title("Tangent")
```
