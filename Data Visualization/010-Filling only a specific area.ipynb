***Filling Area on Line Plots & Filling Only a Specific Area***

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 10, 0.1)
y = x ** 4

plt.figure(figsize= (5, 5))

plt.style.use("seaborn")

plt.plot(x, y)
```

```python
x = np.arange(0, 10, 0.1)
y = x ** 4

plt.style.use("seaborn")

plt.figure(figsize= (5, 5))

plt.plot(x, y)

plt.fill_between(x, y, np.max(x), color= "b", alpha= 0.3)
```
