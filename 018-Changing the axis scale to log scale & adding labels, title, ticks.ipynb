```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = [189, 185, 195, 149, 189, 147, 154, 174, 169, 195, 159, 192, 155, 191, 153, 157, 140, 144, 172, 157, 181, 182, 166, 167, 140, 130, 150,
        190, 171, 189, 142, 153, 159, 161, 169, 165, 133, 188]
```

```python
plt.figure(figsize= (5, 5))

plt.hist(data, bins= 21, align= "right", color= "c", edgecolor= "r", linewidth= 1)
```

```python
plt.figure(figsize= (5, 5))

plt.hist(data, bins= range(140, 180, 5), align= "left", color= "c", edgecolor= "r", linewidth= 1)
```

```python
plt.hist(data, bins= range(140, 170, 1), align= "left", color= "white", edgecolor= "r", linewidth= 1)

plt.xticks(range(140, 170, 1))
```

```python
plt.hist(data, bins= range(140, 155, 1), align= "left", color= "b", edgecolor= "r", linewidth= 1)

plt.xticks(range(140, 155, 1))
```

```python
plt.hist(data, bins= range(140, 155, 1), align= "left", color= "g", edgecolor= "r", linewidth= 1)

plt.xticks(range(140, 155, 1))
plt.yticks(range(1, 10, 1))
```

```python
plt.hist(data, bins= range(140, 155, 1), align= "left", color= "y", edgecolor= "r", linewidth= 1)

plt.xticks(range(140, 155, 1))
plt.yticks(range(1, 5, 1))
```

```python
plt.hist(data, bins= range(140, 155, 1), align= "left", color= "m", edgecolor= "r", linewidth= 1)

plt.xticks(range(140, 155, 1))
plt.yticks(range(1, 4, 1))

plt.title("Example of Histogram Plot")
plt.xlabel("Students Ages in cm")
plt.ylabel("Students Quantity")
```
