```python
def fillna_age (cols):
    
    Age= cols[0]
    Pclass= cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 38
            
        elif Pclass == 2:
            return 29
            
        else:
            return 25
            
    else:
        return Age
```

```python
df_taitanic["Age"] = df_taitanic[["Age", "Pclass"]].apply(fillna_age, axis= 1)
```

```python
sns.heatmap(df_taitanic.isnull(), yticklabels= False, cbar= True, cmap= "Blues")
```

```python
df_taitanic.drop(["PassengerId", "Sex", "Embarked", "Name", "Ticket", "Pclass"], axis= 1, inplace= True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_taitanic.drop("Survived", axis= 1)
y = df_taitanic["Survived"]
```

```python
X
```

```python
X.columns
```

```python
X.columns = X.columns.astype(str)
```

```python
X.columns
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(X_train,y_train)
```

```python
Prediction = LR.predict(X_test)
Prediction
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, Prediction) 
```
