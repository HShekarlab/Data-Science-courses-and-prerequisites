***Getting our first basic line plot***

```python
# from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 4, 9, 16, 25, 36]
y = [1, 2, 3, 4, 5, 6]
```

```python
plt.plot(x, y)
```

***Changing Figure Size and Saving the Figure***

```python
f = plt.figure()

f.set_figwidth(5)
f.set_figheight(2)

plt.plot(x, y)
```

```python
plt.figure(figsize= (2, 5))
display(plt.plot(x, y))
```
