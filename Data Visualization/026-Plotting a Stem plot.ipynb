```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 6, 7, 3, 8]
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y)
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= "-")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= "--")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= "-.")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "r+")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "ro")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "m*")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= "g:", 
         markerfmt= "m*")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "m*", 
         bottom= 1)
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "c*", 
         bottom= 2)
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "co", 
         bottom= 3)
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "yo", 
         bottom= 4)
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "m*", 
         bottom= 1, 
         basefmt= "g")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "go", 
         bottom= 1, 
         basefmt= "g", 
         label= "Python")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "mo", 
         bottom= 1, 
         basefmt= "g", 
         label= "Python")

plt.legend()
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "c*", 
         bottom= 1, 
         basefmt= "g", 
         label= "Python", 
         use_line_collection= True)

plt.legend()
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "m*", 
         bottom= 1, 
         basefmt= "g", 
         label= "Python", 
         use_line_collection= True,
         orientation= "horizontal")

plt.legend()
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y,
         linefmt= ":", 
         markerfmt= "g*", 
         bottom= 1, 
         basefmt= "g", 
         label= "Python", 
         use_line_collection= True)

plt.legend(loc= "lower right")
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "bo", 
         bottom= 1, 
         basefmt= "g", 
         label= "Python", 
         use_line_collection= True)

plt.legend(loc= (0.7, 0.5))
```

```python
plt.figure(figsize= (4, 4))

plt.stem(x, y, 
         linefmt= ":", 
         markerfmt= "r*", 
         bottom= 1, 
         basefmt= "g", 
         label= "Python", 
         use_line_collection= True)

plt.legend(loc= (0.7, 0.3))
```
