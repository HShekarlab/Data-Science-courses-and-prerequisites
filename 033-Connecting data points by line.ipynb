```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 6, 8, 10]
```

```python
plt.figure(figsize= (5, 5))

plt.plot(x_values, y_values, "o-", label= "Data")

plt.title("Title of the Plot")
plt.xlabel("X - axis label")
plt.ylabel("Y - axis label")

plt.legend()
```

```python
plt.figure(figsize= (5, 5))

plt.plot(x_values, y_values, "*-", label= "Data")

plt.title("Title of the Plot")
plt.xlabel("X - axis label")
plt.ylabel("Y - axis label")

plt.legend()
```

```python
plt.figure(figsize= (5, 5))

plt.plot(x_values, y_values, "^--", label= "Data")

plt.title("Title of the Plot")
plt.xlabel("X - axis label")
plt.ylabel("Y - axis label")

plt.legend()
```

```python
plt.figure(figsize= (5, 5))

plt.plot(x_values, y_values, "*--", label= "Data")

plt.title("Title of the Plot")
plt.xlabel("X - axis label")
plt.ylabel("Y - axis label")

plt.legend()
```

```python
plt.figure(figsize= (5, 5))

plt.plot(x_values, y_values, "-.", label= "Data")

plt.title("Title of the Plot")
plt.xlabel("X - axis label")
plt.ylabel("Y - axis label")

plt.legend()
```
