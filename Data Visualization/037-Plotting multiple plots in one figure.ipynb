```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y_1 = np.sin(x)
y_1
```

```python
y_2 = np.cos(x)
y_2
```

```python
y_3 = np.tan(x)
y_3
```

```python
fig, axes = plt.subplots(1, 3, figsize= (10, 4))

print(fig)
```

```python
fig, axes = plt.subplots(1, 3, figsize= (10, 4))

print(axes)
```

```python
fig, axes = plt.subplots(1, 3, figsize= (10, 4))

axes[0].plot(x, y_1)
axes[0].set_title("Sin")

axes[1].plot(x, y_2)
axes[1].set_title("Cosin")

axes[2].plot(x, y_3)
axes[2].set_title("Tangent")

fig.suptitle("Trigonometric Functions")
```

```python
fig, axes = plt.subplots(1, 3, figsize= (10, 4))

axes[0].plot(x, y_1, "b*")
axes[0].set_title("Sin")

axes[1].plot(x, y_2, "m--")
axes[1].set_title("Cosin")

axes[2].plot(x, y_3, "c-.")
axes[2].set_title("Tangent")

fig.suptitle("Trigonometric Functions")
```
