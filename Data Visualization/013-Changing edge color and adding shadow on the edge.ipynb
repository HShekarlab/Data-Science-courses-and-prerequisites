```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([35, 20, 25, 10, 15])
my_labels = ["Strawberry", "Mango", "Cherry", "Avocado", "Nectarine"]
colors = ["red", "cyan", "navy", "deeppink", "yellow"]

plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels)
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, wedgeprops= {'linewidth': 2, "edgecolor": "white"})
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, wedgeprops= {'linewidth': 1, "edgecolor": "white"})
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, wedgeprops= {'linewidth': 4, "edgecolor": "white"})
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, wedgeprops= {'linewidth': 2, "edgecolor": "white"}, colors= colors)
```

```python
plt.figure(figsize= (5, 5))

plt.pie(x, labels= my_labels, wedgeprops= {'linewidth': 2, "edgecolor": "white"}, colors= colors, shadow= True)
```
