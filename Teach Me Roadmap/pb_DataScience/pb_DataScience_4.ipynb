```python
def Doller_Remove(x):
    return x.replace("$", "")
```

```python
txt_2 = "50$"
Doller_Remove(txt_2)
```

```python
def K_Remove(x):
    return x.replace("K", "")
```

```python
txt_3 = "K50K"
K_Remove(txt_3)
```

```python
df["Salary Estimate"] = df["Salary Estimate"].apply(Doller_Remove)
df["Salary Estimate"]
```

```python
df["Salary Estimate"] = df["Salary Estimate"].apply(K_Remove)
df["Salary Estimate"]
```
