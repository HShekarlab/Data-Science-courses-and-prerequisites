```python
df = df.drop(["BoosterVersion", "Serial", "Longitude", "Latitude"], axis= 1)
```

```python
df.info()
```

```python
sns.catplot(data= df, x= "FlightNumber", y= "PayloadMass", hue= "Class", aspect= 5)
plt.xlabel("Flight Number", fontsize=20)
plt.ylabel("Pay load Mass (kg)", fontsize=20)
```

```python
sns.catplot(data= df, x= "FlightNumber", y= "LaunchSite", hue= "Class", aspect= 5)
plt.xlabel("Flight Number", fontsize= 20)
plt.ylabel("Pay load Mass (kg)", fontsize= 20)
```

```python
sns.catplot(data= df, x= "LaunchSite", y= "PayloadMass", hue= "Class", aspect= 5)
plt.xlabel("Launch Site", fontsize= 20)
plt.ylabel("Pay load Mass (kg)", fontsize= 20)
```

```python
sns.catplot(data= df, x= "FlightNumber", y= "Orbit", hue= "LaunchSite", aspect= 5)
plt.xlabel("FlightNumber", fontsize= 20)
plt.ylabel("Orbit", fontsize= 20)
```
