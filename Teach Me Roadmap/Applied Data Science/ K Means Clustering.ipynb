```python
k = 1
```

```python
kmeans = KMeans(n_clusters= k)
```

```python
kmeans.fit(df)
```

```python
kmeans.inertia_
```

```python
sse = {}
```

```python
for k in range(1, 10):
    kmeans = KMeans(n_clusters= k)
    kmeans.fit(df)
    sse[k] = kmeans.inertia_
```

```python
sse
```

```python
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
```
