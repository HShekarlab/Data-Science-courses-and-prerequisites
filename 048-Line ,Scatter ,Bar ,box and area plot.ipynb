***This will enable the use of Cufflinks with Plotly in offline mode***

```python
import pandas as pd
import plotly.express as px
import cufflinks as cf
cf.go_offline()

# import plotly.offline as pyo
# import plotly.graph_objs as go
```

***Line Plot***

```python
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv")
df
```

```python
fig = px.line(df, x= "Date", y= "AAPL.High", title= "AAPL High Price")
fig.show()
```

***Scatter Plot***

```python
fig = px.scatter(df, x= "Date", y= "AAPL.High", title= "AAPL High Price")
fig.show()
```

***Bar Plot***

```python
fig = px.bar(df, x= "Date", y= "AAPL.High", title= "AAPL High Price")
fig.show()
```

```python
fig = px.bar(df, x= "Date", y= "AAPL.High", title= "AAPL High Price", template= "plotly_dark")
fig.show()
```

***Box Plot & Area Plot***

```python
fig = px.box(df, x= "Date", y= "AAPL.High", title= "AAPL High Price")
fig.show()
```

```python
fig = px.box(df, x= "Date", y= "AAPL.High", title= "AAPL High Price", template= "plotly_dark")
fig.show()
```

```python
fig = px.area(df, x= "Date", y= "AAPL.High", title= "AAPL High Price")
fig.show()
```

```python
fig = px.area(df, x= "Date", y= "AAPL.High", title= "AAPL High Price", template= "plotly_dark")
fig.show()
```

```python
data = pd.DataFrame({"x": [1, 2, 3, 4, 5],
                     "y_1": [2, 4, 1, 5, 3],
                     "y_2": [1, 3, 2, 4, 2]})

fig = px.area(data, x= "x", y= ["y_1", "y_2"], title= "Area Plot")
fig.show()
```
