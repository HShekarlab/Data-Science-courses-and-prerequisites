```python
landing_outcomes = df["Outcome"].value_counts()
landing_outcomes
```

```python
for i, outcome in enumerate(landing_outcomes.keys()):
    print(i, outcome)
```

```python
bad_outcomes = set(landing_outcomes.keys()[[1, 3, 5, 6, 7]])
```

```python
bad_outcomes
```

```python
bad_outcomes_1 = set(["False ASDS", "False Ocean", "False RTLS", "None ASDS", "None None"])
```

```python
bad_outcomes_1
```

```python
def onehot(item):
    if item in bad_outcomes:
        return 0
    else:
        return 1

landing_class = df["Outcome"].apply(onehot)
landing_class
```

```python
df["Class"] = landing_class
df[["Class"]].head(12)
```

```python
successfully_landed = ["True ASDS", "True Ocean", "True RTLS"]
df["Class_1"] = df["Outcome"].map(lambda x : int(x in successfully_landed))
```

```python
df = df.drop("Class_1", axis= 1)
```
