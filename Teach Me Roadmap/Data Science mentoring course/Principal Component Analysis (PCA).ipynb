```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# scaler.fit(df)
# scaled_data = scaler.transform(df)

scaled_data = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
# pca = pca.fit(scaled_data)
# x_pca = pca.transform(scaled_data)

x_pca = pca.fit_transform(scaled_data)
```

```python
x_pca 
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize= (5, 5))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c= cancer["target"], cmap= "plasma")
plt.xlabel("First load_breast_cancer")
plt.ylabel("Second load_breast_cancer")
```

```python
plt.figure(figsize= (5, 5))
sns.scatterplot(x= x_pca[:, 0], y= x_pca[:, 1], hue= cancer["target"], palette= "rainbow")
plt.xlabel("First load_breast_cancer")
plt.ylabel("Second load_breast_cancer")
```

```python
df_feat = pd.DataFrame(cancer["data"], columns= cancer["feature_names"])
df_target = pd.DataFrame(cancer["target"], columns= ["Cancer"])

df_1 = pd.concat([df_feat, df_target], axis= 1)
sns.scatterplot(data= df_1, x= "mean radius", y= "mean texture", hue= "Cancer", palette= "viridis")
```
